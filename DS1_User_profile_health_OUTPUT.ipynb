## ASSESSMENT 1
### To find  (1) Participant level for (i) Season Profile Created, (ii) Latest BMI;


#### Assumptions made
1. Total number of unique participants 5002

2. Removed 2 duplicate IDs

3. If Profilecreation date null, I have considered "UpdatedAT" if both are null then I have consider the latest date[2020-03-12]  for 115 cases out of total 5002 unique participants

4.  Assumed Weight_Latest in KG & Height_Latest in CM

5. If any user id is not available in "DS1_user_health_table", I assumed that user has not participated after profile creation or did not his/her data so I have considered his/her Latest Weight/Height recorded during profile creation.


```python
import pandas as pd
import numpy as np
import re
dfh=pd.read_csv('/Users/nsearch/Desktop/HPB_ASSESMENT/Assessment_Inputs/DS1_user_health_table.csv')
dfp=pd.read_csv('/Users/nsearch/Desktop/HPB_ASSESMENT/Assessment_Inputs/DS1_user_profile_table.csv')
print('Total no of rows available in the DS1_user_health_table',dfh.shape)
print('Total no of rows available in the DS1_user_profile_table',dfp.shape)
dfp=dfp.drop_duplicates(subset=None, keep='first', inplace=False)
#dfp=dfp.set_index('id')
#dfh
```

```python
#Profile created date
df_min=dfh[dfh.groupby('id').UpdateAt.transform('min') == dfh['UpdateAt']].set_index('id')

#latest height and weight
df_max=dfh[dfh.groupby('id').UpdateAt.transform('max') == dfh['UpdateAt']].set_index('id')
#
result = dfp.join(df_min,rsuffix='_Min').join(df_max,rsuffix='_Max')
#result
#
```

####  TO get the profile created  date and recent activity data,latest Height and Weight

```python
result['Min_Date'] = np.nan
result['Min_Date'] = result.Min_Date.fillna(result.ProfileCreateDate).fillna(result.UpdateAt).fillna(result.UpdateAt_Max)

result['Max_Date'] = np.nan
result['Max_Date'] = result.Max_Date.fillna(result.UpdateAt_Max).fillna(result.UpdateAt).fillna(result.ProfileCreateDate)

result['Latest_Weight'] = np.nan
result['Latest_Weight'] = result.Latest_Weight.fillna(result.Weight_Max).fillna(result.Weight_Min).fillna(result.Weight)

result['Latest_Height'] = np.nan
result['Latest_Height'] = result.Latest_Height.fillna(result.Height_Max).fillna(result.Height_Min).fillna(result.Height)
result=result.reset_index()
result = result.drop_duplicates(subset='id', keep="first")
result=result.set_index('id').drop('index',axis=1)
result
```

```python
data=result[[ 'Min_Date', 'Latest_Weight', 'Latest_Height']]
data
```

####  Assume the recent date [2020-03-12] if the profile create date is null to enhance the data quality and it will be ease of processing

```python
data['Profile_created_Date'] = ['2020-03-12T00:00:10.000Z' if x is np.nan else x for x in list(data['Min_Date'])]
data
```

#### Create a profile created season using the given date range

```python
bins_dt = pd.date_range('2015-08-03', freq='364D', periods=6)
#bins_dt
data['Profile_created_Date'] = pd.to_datetime(data['Profile_created_Date'])
data['season_created_profile']=pd.cut(data['Profile_created_Date'],bins=bins_dt,labels=[1,2,3,4,5])
data
```

#### Calculate the BMI using latest Weight and Height 
#### BMI = (Latest_Weight/Latest_Height/Latest_Height)*1000

```python
data['BMI']=np.round((data['Latest_Weight'] / data['Latest_Height']/ data['Latest_Height'])*10000,2)

output=data[['season_created_profile' ,'Profile_created_Date','BMI', 'Latest_Weight','Latest_Height']]
output['BMI']=output['BMI'].replace(np.nan,0)
output
```
