```python
#importing libraries
import pandas as pd
import numpy as np
import re
# read the source data
df=pd.read_csv('/Users/nsearch/Desktop/HPB_ASSESMENT/Assessment_Inputs/DS2_fitness.csv')
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])
df.head()

list_of_activity_date=list(df['ActivityDate'])
#dictionary to convert the  week  number into days
weekday_dict = {1:'Monday', 2: 'Tuesday', 3: 'Wednesday', 4:'Thursday', 5: 'Friday', 6:'Saturday', 7:'Sunday'}
#To get the weeknum, weekdays and day 
weeknum=[]
weekday=[]
day=[]
for i in list_of_activity_date:
    
    val=i.isocalendar()[1] - i.replace(day=1).isocalendar()[1] + 1
    weekdy=i.isoweekday()
    weeknum.append(val)
    weekday.append(weekdy)
    day.append(weekday_dict.get(weekdy))
#create the new columns for weeknum,weekdays and daya   
df['weeknumber']=weeknum
df['weekday']=weekday
df['day']=day
#print(len(df))
df['StepCount'] = np.where(df.StepCount <999, 0, df.StepCount)
#print(len(df))
#pivot table to convert the required output

table = pd.pivot_table(df, values='StepCount', index=['id', 'weeknumber'], columns=['day'], aggfunc=np.sum)
table=table.reset_index()#.replace(np.nan,0)
table=table.replace(0.0,np.nan)
table['average']=np.round(table[['Friday', 'Monday', 'Saturday', 'Sunday', 'Thursday', 'Tuesday', 'Wednesday']].mean(axis=1),2)
#convert into expected Format 
Req_data=table[['id','weeknumber', 'Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday','average' ]].replace(np.nan,0)

col_name=[ 'Monday','Tuesday','Wednesday','Thursday','Friday', 'Saturday', 'Sunday']
new_name=['steps_Usedays_'+i for i in col_name]
Req_data=Req_data.rename(columns=dict(zip(col_name,new_name))).rename(columns={'average': 'Avg_Daily_usedays_Steps','weeknumber':'Challenge_week'}).set_index('id')
Req_data.head()
```
