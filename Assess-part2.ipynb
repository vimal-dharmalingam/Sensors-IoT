```python
def week_number_of_month(date_value):
     return (date_value.isocalendar()[1] - date_value.replace(day=1).isocalendar()[1] + 1)

# from datetime import datetime
# datetime.fromtimestamp()
print(df['ActivityDate'][0].isocalendar()[1] - df['ActivityDate'][0].replace(day=1).isocalendar()[1] + 1)
# MONDAY....
print(df['ActivityDate'][0].isoweekday())

```

```python
alll=list(df['ActivityDate'])
weekday_dict = {1:'Monday', 2: 'Tuesday', 3: 'Wednesday', 4:'Thursday', 5: 'Friday', 6:'Saturday', 7:'Sunday'}
weeknum=[]
weekday=[]
day=[]
for i in alll:
    
    val=i.isocalendar()[1] - i.replace(day=1).isocalendar()[1] + 1
    weekdy=i.isoweekday()
    weeknum.append(val)
    weekday.append(weekdy)
    day.append(weekday_dict.get(weekdy))
    
df['weeknumber']=weeknum
df['weekday']=weekday
df['day']=day
```
