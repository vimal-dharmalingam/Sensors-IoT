```python
dfp.head()
dfp=dfp.rename(columns={'ProfileCreateDate':'UpdateAt'})
dfp
```

```python
s1=pd.concat([dfp,dfh])
s1
```

```python
s1=s1.drop_duplicates(subset=None, keep='first', inplace=False)
s1.shape
```

```python
len(s1['id'].unique())
```

```python
s1.head(30)
s1=s1.dropna()
len(s1['id'].unique())
```

```python


s1['UpdateAt'] = pd.to_datetime(s1['UpdateAt'])
s1
```

```python
df_max = s1[s1.groupby('id').UpdateAt.transform('max') == s1['UpdateAt']]
df_max=df_max.set_index('id')

df_min=s1[s1.groupby('id').UpdateAt.transform('min') == s1['UpdateAt']]
df_min=df_min.set_index('id')


f=pd.concat([df_max,df_min[['UpdateAt']].rename(columns={'UpdateAt':'Min_Date'})],axis=1,sort=False)
f.head(60)
```

```python
bins_dt = pd.date_range('2015-08-03', freq='364D', periods=6)
#bins_dt
f['season_created_profile']=pd.cut(f['Min_Date'],bins=bins_dt,labels=[1,2,3,4,5])
f.head(30)
```

```python
f['BMI']=np.round((f['Weight'] / f['Height']/ f['Height'])*10000,2)
f
req=f[['season_created_profile' ,'BMI', 'Weight','Height']].rename(columns={'Weight':'Latest_Weight','Height':'Latest_Height'})

req.head(50)
```
