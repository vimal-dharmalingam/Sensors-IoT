```python
PARTY_ID = MERCHANT.ID.unique().shape[0]
MERCHANTS= MERCHANT.NAME.unique().shape[0]
PARTY_ID
MERCHANTS
```

```python
mer_id=list(MERCHANT['MERCHANT_NAME'].unique())
cust_id=list(MERCHANT['CUST_ID'])
```

```python
import numpy as np
SPARSE_matrix = np.zeros((PARTY_ID , MERCHANTS))
SPARSE_matrix
```

```python
SPARSE_matrix = np.zeros((PARTY_ID, MERCHANTS))
for line in MERCHANT.itertuples():
    SPARSE_matrix[line[4]-1, line[5]-1] = line[3]
SPARSE_matrix
```

```python
mer= [i.rstrip() for i in mer_id] 
mer
```

```python
df=pd.DataFrame(SPARSE_matrix,columns=mer)
```

```python
# Create a placeholder dataframe listing item vs. item
place_hol = pd.DataFrame(index=df.columns,columns=df.columns)
place_hol
```

```python
%timeit 
from scipy.spatial.distance import cosine
# Lets fill in those empty spaces with cosine similarities
# Loop  through the columns
for i in range(0,len(place_hol.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(place_hol.columns)) :
      # Fill in placeholder with cosine similarities
      place_hol.ix[i,j] = 1-cosine(df.iloc[:,i],df.iloc[:,j])
```

```python
# Create a placeholder merchants for closes neighbours to an item
merchant_neighbours = pd.DataFrame(index=place_hol.columns,columns=range(1,11))
 
# Loop through our similarity dataframe and fill in neighbouring merchnat names
for i in range(0,len(place_hol.columns)):
    merchant_neighbours.ix[i,:10] = place_hol.ix[0:,i].sort_values(ascending=False)[:10].index
```
